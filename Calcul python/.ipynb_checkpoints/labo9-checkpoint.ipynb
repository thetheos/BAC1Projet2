## Generate Plotly

```python

data = []
data.append(go.Scatter(x=df_test.time,y=df_test.chanA,name="Channel A"))
data.append(go.Scatter(x=df_test.time,y=df_test.chanB,name="Channel B"))


plotly.offline.plot({
    "data": data,
    "layout": go.Layout(title="Pico_Plotter", xaxis=dict(title="Time (¬µs)"), yaxis=dict(title="Volatage (V)")),   
}, filename= "test.html")
```
